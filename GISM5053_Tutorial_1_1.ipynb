## 3.1 Create list

```python
# two ways to create empty list
lst_a = list()
lst_b = [] # Square brackets
print(lst_a,lst_b)
```

```python
# then append elements into list
# different types of data can be stored in the same list 
lst_a.append(1)
lst_a.append('hello')
lst_a.append(True)
print(lst_a)
```

```python
# create an non-empty list
lst_c = [3,4,5,6,7,8]
print(lst_c)
```

## 3.2 Index of list

```python
# index of list
print(lst_a[0]) # firt element. Note: Index in Python is counted from 0
```

```python
print(lst_a[1:2]) # from the sencond to the third element(excluded)
```

```python
print(lst_c[-1])
print(lst_c[-3:-1])
print(lst_c[-3:])
```

# 9 Deal with errors

```python
lst_d = [1,2,3,4,'wtf',5]

for i in lst_d:
  print(i)
```

```python
lst_d = [1,2,3,4,'wtf',5]
for i in lst_d:
  print(i+1)
```

```python
lst_d = [1,2,3,4,'wtf',5]
for i in lst_d:
  try:
    print(i+1)
  except: #if try something meets error,...
    print('the error element is:', i)
```
